```python
import pandas as pd
filename = 'beneficiarios.csv'
data = pd.read_csv(filename)
print(data)
```

```python
# Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas.
print('Dimensiones:', data.shape)
print('Número de elemntos:', data.size)
print('Nombres de columnas:', data.columns)
print('Nombres de filas:', data.index)
print('Tipos de datos:\n', data.dtypes)
print('Primeras 10 filas:\n', data.head(10))
print('Últimas 10 filas:\n', data.tail(10))
```

```python
# Mostrar por pantalla las filas pares del DataFrame.
print(data.iloc[range(0,data.shape[0],2)])
```
